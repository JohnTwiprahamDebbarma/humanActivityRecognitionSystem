```python
# plotting X_train (3d plot) 6 columns
def plot_3d(X,activity):
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=X[:, 3], cmap='brg')
    ax.set_xlabel('a1')
    ax.set_ylabel('a2')
    ax.set_zlabel('a3')
    ax.set_title(f"{activity}")
    plt.show()
```

```python
X_train_activity = {}
activities = ["LAYING","SITTING","STANDING","WALKING","WALKING_DOWNSTAIRS","WALKING_UPSTAIRS"]
for i in range(len(activities)):
    X_train_activity[activities[i]] = X_train[i*(21):(i+1)*21]
X_train_activity["LAYING"].shape
```

```python
# plot 3d plots for each activity in 6 columns
for activity in activities:
    plot_3d(X_train_activity[activity],activity)
```
